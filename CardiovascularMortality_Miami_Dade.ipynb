```python
import geopandas as gpd
gdf = gpd.read_file('CardiovascularMortality2010_gdb_8005841059351655123.geojson')
gdf.head()
```

```python
print(gdf.columns)
```

```python
gdf.plot(column='DEATHCOUNT', cmap='Reds', legend=True, figsize=(12, 8))
```

```python
import folium

# Create the base map centered around Miami
m = folium.Map(location=[25.76, -80.19], zoom_start=10)

# Add choropleth layer
folium.Choropleth(
    geo_data=gdf,              # GeoDataFrame
    data=gdf,                  # DataFrame with the values
    columns=['ZIPCODE', 'DEATHCOUNT'],  # Must match exact column names
    key_on='feature.properties.ZIPCODE',  # Must match the GeoJSON properties
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Death Count per ZIP Code'
).add_to(m)

m
```
